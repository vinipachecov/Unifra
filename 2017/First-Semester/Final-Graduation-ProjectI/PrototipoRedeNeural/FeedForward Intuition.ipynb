```python
A = np.random.randn(100,5)
```

# One hunderd samples in 5 classes (k =5)

```python
A
# Remember we will have more than one sample, lets say a 100 
# now we have to compute them all together, let's try the same way we did before
```

```python
expA = np.exp(A)
##Calculate de exp of all samples at the same time
```

```python
expA
```

```python
answer = expA / expA.sum()
## this is wrong. 
#Remember expA is a matrix, not a vector of one sample. 
# In this situation we need to take the sum for every row and divide each element of the row for this sum.
# Remember that exp(a1) / exp(a1) + exp(a2) + exp(a3) + exp(a4) + exp(a5)....exp(ak) !!!
```

# î is WRONG

```python
answer = expA / expA.sum(axis=1)
##This seems right, but won't work because expA.sum(axis=1) returns a (N,) shape, which will not allow division.
```

```python
answer = expA / expA.sum(axis=1, keepdims=True)
##By using keepdmins=True attribute, our expA.sum() will return our (N,1) matrix (each row sum of exponentials) 
## and allow us to make the division
```

```python
answer
```

```python
answer.sum(axis=1)
##check if the sum of every row of our answer is 1 as we are talking about probabilities
```

```python
expA.sum(axis=1, keepdims=True)
##See the value of how much we are dividing every row
```
